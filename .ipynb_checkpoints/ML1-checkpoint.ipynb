```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Wine.csv')
```

```python
df.keys()
```

```python
df.head(5)
```

```python
df.Customer_Segment.unique()
```

```python
print(df.isnull().sum())
```

```python
x = df.drop('Customer_Segment', axis=1)
y = df['Customer_Segment']
```

```python
for col in x.columns:
    sc = StandardScaler()
    x[col] = sc.fit_transform(x[[col]])
```

```python
x.head(5)
```

```python
pca = PCA()
x_pca = pca.fit_transform(x)
```

```python
explained_variance_ratio = pca.explained_variance_ratio_
```

```python
plt.plot(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio.cumsum(), marker='o', linestyle='--')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Explained Variance Ratio')
plt.show()
```

```python
n_components = 12
pca = PCA(n_components = n_components)
x_pca = pca.fit_transform(x)
```

```python
x_pca.shape
```

```python
red_indices = y[y == 1].index
white_indices = y[y == 2].index

plt.scatter(x_pca[red_indices, 0], x_pca[red_indices, 1], c='red', label='Red Wine')
plt.scatter(x_pca[white_indices, 0], x_pca[white_indices, 1], c='green', label='White Wine')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.title('PCA: Red Wine vs White Wine')
plt.show()
```
